```python
c1 = 0
c2 = 0
c3 = 0
c4 = 0
for i in f['Y']:
    if (i[0] == 1.0) and (i[1] == i[2] == i[3] == 0.0):
        c1 += 1
    elif (i[1] == 1.0) and (i[0] == i[2] == i[3] == 0.0):
        c2 += 1
    elif (i[2] == 1.0) and (i[0] == i[1] == i[3] == 0.0):
        c3 += 1
    elif (i[3] == 1.0) and (i[1] == i[2] == i[0] == 0.0):
        c4 += 1
print('class 1: ', c1)
print('class 2: ', c2)
print('class 3: ', c3)
print('class 4: ', c4)
```
